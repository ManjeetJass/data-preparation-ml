```python
# Program to visualize missing values in dataset

# Importing the libraries
import pandas as pd
import missingno as msno
```

```python
# Loading the dataset
df = pd.read_csv("titanic-test.csv")
```

```python
# Visualize missing values as a matrix 
msno.matrix(df) 
```

```python
msno.bar(df)
```

<h3><b>Delete Column

```python
# Delete column "Cabin"
df_less_missing_rows = df.drop(['Cabin'], axis=1)
msno.bar(df_less_missing_rows)
```

<h3><b> Replace missing values

```python
# replace missing values with the median.
med = df_less_missing_rows['Age'].median()
print(med)
df_less_missing_rows['Age'] = df_less_missing_rows['Age'].fillna(med)
```

```python
msno.bar(df_less_missing_rows)
```

```python
# replace missing values with the most frequent.
most = df_less_missing_rows['Fare'].mode()
print(most.get(key = 0))
df_less_missing_rows['Fare'] = df_less_missing_rows['Fare'].fillna(most.get(key = 0))
```

```python
msno.bar(df_less_missing_rows)
```
